```python
from itertools import combinations
```

```python
def calculate_win_rate(group):
    total_matches = len(group)
    wins = group['win'].sum()
    return wins / total_matches
```

```python
grouped = df.groupby('match_id')

# Dictionary to hold win rates for each hero combination
combination_win_rates = {}
```

```python
for match_id, group in grouped:
    heroes = group['hero_name'].tolist()
    win_rates = group.groupby('hero_name')['win'].mean()

    # Generate all possible combinations of heroes for this match
    for combo_size in range(2, len(heroes) + 1):  # From pairs to all heroes
        for combo in combinations(heroes, combo_size):
            # Create a combination key
            combo_key = tuple(sorted(combo))
            if combo_key not in combination_win_rates:
                combination_win_rates[combo_key] = []
            # Calculate win rate for this combination
            combo_win_rate = calculate_win_rate(group[group['hero_name'].isin(combo)])
            combination_win_rates[combo_key].append(combo_win_rate)
```

```python
average_win_rates = {combo: sum(win_rates) / len(win_rates) for combo, win_rates in combination_win_rates.items()}

# Step 4: Find top combinations with highest average win rate
top_combinations = sorted(average_win_rates.items(), key=lambda x: x[1], reverse=True)

# Display the top combinations
print("Top Hero Combinations by Win Rate:")
for combo, win_rate in top_combinations[:15]:  # Display top 10 combinations
    print(f"Combination: {combo}, Average Win Rate: {win_rate:.2f}")
```

```python
hero_stats = df.groupby('hero_name').agg({
    'win': 'mean',  # Win rate
    'kills': 'mean',
    'deaths': 'mean',
    'assists': 'mean',
    'gold_per_min': 'mean',
    'xp_per_min': 'mean',
    'hero_damage': 'mean'
}).reset_index()

# Compute KDA (Kills + Assists) / Deaths
hero_stats['kda'] = (hero_stats['kills'] + hero_stats['assists']) / hero_stats['deaths']
```

```python
metrics = ['win', 'kda', 'hero_damage', 'gold_per_min', 'xp_per_min']

for metric in metrics:
    min_value = hero_stats[metric].min()
    max_value = hero_stats[metric].max()
    hero_stats[f'{metric}_norm'] = (hero_stats[metric] - min_value) / (max_value - min_value)
```

```python
hero_stats['score'] = hero_stats[['win_norm', 'kda_norm', 'hero_damage_norm', 'gold_per_min_norm', 'xp_per_min_norm']].mean(axis=1)
```

```python
# Sort heroes by their total score
hero_stats_sorted = hero_stats.sort_values(by='score', ascending=False)

# Select top 15 and bottom 15 heroes
top_15_heroes = hero_stats_sorted.head(15)
bottom_15_heroes = hero_stats_sorted.tail(15)

# Display top 15 heroes
print("Top 15 Heroes:")
print(top_15_heroes[['hero_name', 'score', 'win', 'kda', 'hero_damage', 'gold_per_min', 'xp_per_min']])

# Display bottom 15 heroes
print("\nBottom 15 Heroes:")
print(bottom_15_heroes[['hero_name', 'score', 'win', 'kda', 'hero_damage', 'gold_per_min', 'xp_per_min']])
```
