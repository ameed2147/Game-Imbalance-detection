```python
import requests
import sqlite3
import time
import json

# Define the base URL and endpoints
base_url = "https://api.opendota.com/api"
match_endpoint = "/matches/"
player_recent_matches_endpoint = "/players/{account_id}/recentMatches"
```

```python
# Function to get match details
def get_match_details(match_id):
    url = base_url + match_endpoint + str(match_id)
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None
```

```python
# Initialize SQLite database
conn = sqlite3.connect('dota_data.db')
c = conn.cursor()
```

```python
c.execute('''CREATE TABLE IF NOT EXISTS matches
             (match_id INTEGER PRIMARY KEY, data TEXT)''')
```

```python
def save_match_details(match_data):
    if match_data:
        match_id = match_data['match_id']
        data = json.dumps(match_data)
        c.execute("INSERT OR IGNORE INTO matches (match_id, data) VALUES (?, ?)", (match_id, data))
        conn.commit()
```

```python
# Function to collect a large number of matches
def collect_matches(start_match_id, count):
    collected = 0
    match_id = start_match_id

    while collected < count:
        match_data = get_match_details(match_id)
        if match_data:
            save_match_details(match_data)
            collected += 1
            print(f"Collected match {match_id}")
        else:
            print(f"Failed to collect match {match_id}")
        match_id += 1
        time.sleep(1)  # To avoid hitting rate limits
```

```python
import sqlite3
import json
import pandas as pd

# Function to connect to the database
def connect_db(db_name):
    conn = sqlite3.connect(db_name)
    return conn
```

```python
# Function to fetch match details
def fetch_match_details(conn):
    c = conn.cursor()
    c.execute("SELECT * FROM matches")
    rows = c.fetchall()
    return rows
```

```python
# Function to load data into a DataFrame
def load_data_into_dataframe(rows):
    # Convert rows to a list of dictionaries
    data = [json.loads(row[1]) for row in rows]
    # Create a DataFrame
    df = pd.DataFrame(data)
    return df
```

```python
# Connect to the database
conn = connect_db('dota_data.db')

# Fetch match details
rows = fetch_match_details(conn)

# Load data into a DataFrame
df = load_data_into_dataframe(rows)

# Close the database connection
conn.close()
```
